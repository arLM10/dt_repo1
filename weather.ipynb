```python
from fastapi import APIRouter

router = APIRouter()

@router.get("/api/weather/current")
def get_current_weather(city: str):
    # Mock weather response
    return {
        "city": city,
        "temperature": "22째C",
        "condition": "Cloudy",
        "humidity": "60%"
    }


@router.get("/api/weather/forecast")
def get_weather_forecast(city: str, days: int = 3):
    # Mock 3-day forecast data
    forecast_data = [
        {"day": 1, "temperature": "21째C", "condition": "Sunny"},
        {"day": 2, "temperature": "20째C", "condition": "Rainy"},
        {"day": 3, "temperature": "18째C", "condition": "Cloudy"},
    ]

    return {
        "city": city,
        "days_requested": days,
        "forecast": forecast_data[:days]
    }
```
