```python
from fastapi import APIRouter

router = APIRouter()

# Mock student records
students_db = {
    "12345": {"name": "Aryan Sharma", "course": "Computer Science", "year": 3, "gpa": 8.5},
    "54321": {"name": "Fatma Said", "course": "Mechanical Engineering", "year": 2, "gpa": 8.9},
}


@router.get("/api/student")
def get_student(id: str):
    student = students_db.get(id)

    if student:
        return student
    return {"error": f"No student found with id = {id}"}


@router.get("/api/students/count")
def get_total_students():
    return {
        "students_count": len(students_db)
    }
```
